```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/aadhy/Documents/5th Sem/ML/assignment/tested.csv")
```

```python
df
```

```python
df['Age'].replace(np.nan, df['Age'].mean(), inplace=True)
df['Fare'].replace(np.nan, df['Fare'].mean(), inplace=True)
df.drop('Cabin', axis=1,inplace=True)
df.isna().sum()
```

```python
ax = sns.countplot(x=df['Survived'], palette='RdBu')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
cat_var = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
num_var = ['Age', 'Fare']
```

```python
for column in cat_var:   
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    ax = sns.countplot(x=column, data=df, palette='Accent')
    ax.bar_label(ax.containers[0])
    plt.show()
```

```python
for column in cat_var:
    plt.subplot(1,2,2)
    ax = sns.countplot(x=column, data=df, hue='Survived', palette='Accent')
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    plt.show()
```

```python
df[num_var].describe()
```

```python
for column in num_var:
    plt.figure(figsize=(5,3))
    plt.subplot(1,2,1)
    ax = sns.boxplot(df[column])
    plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Pclass')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='viridis')
plt.title('Correlation')
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title('Number of people Survived')
plt.xlabel('sex')
plt.ylabel('survived')
sns.despine()
```

```python
plt.scatter(df.SibSp,df.Survived)
plt.title("Survived")
plt.xlabel('Sibsp')
plt.ylabel('Survived')
sns.despine()
plt.show()
```
